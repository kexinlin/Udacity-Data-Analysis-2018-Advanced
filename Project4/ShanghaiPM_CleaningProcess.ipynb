### 上海 PM2.5 数据集 数据清洗

```python
# 导入 pandas 库
import pandas as pd
```

```python
# 导入原始数据集
df = pd.read_csv('ShanghaiPM20100101_20151231.csv')
```

```python
# 查看数据集前五行
df.head()
```

```python
# 查看数据变量的类型和不为NaN值的数量
df.info()
```

```python
# 由于数据集存在数据缺失问题，特别是PM这一关键信息，因此删除所有包含缺失值的观察值
df = df.dropna()
df.head()
```

```python
# 重新设置index
df = df.reset_index(drop=True)
df.head()
```

```python
# 保存清理之后的DataFrame
df.to_csv('ShanghaiPM_cleaned.csv', index=False)
```
